```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load adjective sentiment analysis data
df = pd.read_csv("adjective_sentiment_analysis.csv")

# Step 3: Cross-Variation Analysis
variation_sentiment = df.groupby(['Variation'])[['Positive', 'Negative', 'Neutral']].mean().reset_index()

# Visualize sentiment shifts across variations
plt.figure(figsize=(10,6))
sns.lineplot(data=variation_sentiment.melt(id_vars='Variation', var_name='Sentiment', value_name='Average'),
             x='Variation', y='Average', hue='Sentiment', marker='o')
plt.title('Sentiment Shifts Across Scenario Variations')
plt.ylabel('Average Adjective Count')
plt.xlabel('Scenario Variation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("sentiment_shifts_across_variations.pdf")
plt.show()

# Step 4: Cross-Model Comparison
model_sentiment = df.groupby(['Model', 'Character'])[['Positive', 'Negative', 'Neutral']].mean().reset_index()

# Identify most and least biased models
model_pivot = model_sentiment.pivot(index='Model', columns='Character', values='Positive')
model_pivot['Positive Difference'] = abs(model_pivot['Name 1'] - model_pivot['Name 2'])
model_pivot['Positive Bias Ratio'] = model_pivot['Positive Difference'] / (model_pivot['Name 1'] + model_pivot['Name 2'])
model_pivot['Favored Character'] = model_pivot.apply(lambda row: 'Name 1' if row['Name 1'] > row['Name 2'] else 'Name 2', axis=1)

most_biased_model = model_pivot['Positive Difference'].idxmax()
least_biased_model = model_pivot['Positive Difference'].idxmin()

print("Most Biased Model:", most_biased_model)
print("Least Biased Model:", least_biased_model)

# Save detailed model comparison
model_sentiment.to_csv("model_sentiment_comparison.csv", index=False)
model_pivot.to_csv("model_bias_metrics.csv")

# Visualize model bias
plt.figure(figsize=(10,6))
sns.barplot(data=model_sentiment, x='Model', y='Positive', hue='Character')
plt.title('Positive Adjective Usage Across Models and Characters')
plt.ylabel('Average Positive Adjective Count')
plt.xlabel('LLM Model')
plt.xticks(rotation=45)
plt.legend(title='Character')
plt.tight_layout()
plt.savefig("positive_adjective_usage_across_models.pdf")
plt.show()

# Step 5: Lexical Richness Bias (optional)
adjective_df = pd.read_csv("adjective_summary_analysis.csv")
adjective_df['Lexical Richness'] = adjective_df['Unique_Adjectives'] / adjective_df['Total_Adjectives']
richness_summary = adjective_df.groupby(['Model', 'Character'])['Lexical Richness'].mean().reset_index()
richness_summary.to_csv("lexical_richness_by_character.csv", index=False)

# Step 6: Variation Sentiment Shift Analysis
variation_shift = variation_sentiment.set_index('Variation').diff().dropna()
variation_shift.to_csv("variation_sentiment_shifts.csv")
```

```python
import pandas as pd

# Load CSV
file_path = "variation_sentiment_shifts.csv"
df = pd.read_csv(file_path)

# Function to create Overleaf-compatible LaTeX table
def create_overleaf_table(dataframe):
    latex_table = dataframe.to_latex(index=False, float_format="%.4f", escape=True)
    with open("overleaf_sentiment_shifts.tex", "w") as file:
        file.write(latex_table)
    print(latex_table)

# Generate LaTeX table
create_overleaf_table(df)
```
